```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.assembly.go.kr/assm/memact/congressman/memCond/memCondListAjax.do'
params = {'currentPage':1, 'rowPerPage':300}
html = requests.get(url, params=params).text
soup = BeautifulSoup(html, 'html.parser')
for idx, tag in enumerate(soup.select('a[href*=jsMemPop]'),1):
    print('[{}] {}'.format(idx,tag.text))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.family'] = 'Malgun Gothic'

names =[]
for tag in soup.select('a[href*=jsMemPop]'):
    names.append(tag.text)

import pandas as pd
series = pd.Series(names)
series.str.slice(0,1).value_counts().plot(kind='bar', figsize=(12,3))
```

## 자바스크립트 렌더링 크롤링

```python
import re

url = 'https://askdjango.github.io/lv3/'
html = requests.get(url).text
print(html)
```

```python
matched = re.search(r'var s2_courses = (.*?);', html, re.S)

# matched 변수에 저장을 합니다
# '?'를 사용하여 최소 매칭을 해줍니다.
# .: 모든
# *:문자열
# 즉,   .*: 모든 문자열
# re.S를 사용하여 개행을 포함하여 인식한다.
# re.search(r'var courses =();') # 끝에 있는 ()부분을 뽑아내겠다는 뜻
# group으로 모두 출력하는데 1번째줄부터 출력합니다.
# 만약 (0)을 주면 첫번째 줄인 var s2_courses = [도 출력될 것입니다
```

```python
print(matched.group(1))
```

```python
matched = re.search(r'var s2_courses = (.*?);', html, re.S)
# 결과 값을 json_string으로 변경
json_string = matched.group(1)
json.loads(json_string)
```

```python
# 최종 코드
url = 'https://askdjango.github.io/lv3/'
html = requests.get(url).text

# 1,2는 시즌을 의미
matched2 = re.search(r'var s2_courses = (.*?);', html, re.S)
json_string2 = matched2.group(1)

matched1 = re.search(r'var s2_courses = (.*?);', html, re.S)
json_string1 = matched1.group(1)

# json을 output_list에 삽입
output_list2 = json.loads(json_string2)
output_list1 = json.loads(json_string1)

# json 차례대로
for output in output_list2:
    print(output['name'], output['url'])
for output in output_list1:
    print(output['name'], output['url'])
```
