```python
dict = {'col1':[1,11],'col2':[2,22],'col3':[3,33]}
df_dict = pd.DataFrame(dict)
df_dict
```

```python
# 많은 머신러닝 패키지가 기본 데이터 형으로 넘파이 ndarray를 사용한다.
# DataFrame을 넘파이 ndarray로 변환하는 것은 DataFrame 객체의 values를 이용한다.
array3 = df_dict.values
print(type(array3),'\n',array3.shape)
array3
```

```python
# DataFrame을 리스트로 변환
list3 = df_dict.values.tolist() # 리스트로의 변환은 values로 얻은 ndarray에 tolist() 호출
print(list3)

# DataFrame을 딕셔너리로 변환
dict3 = df_dict.to_dict('list') # 딕셔너리로의 변환은 DataFrame 객체의 to_dict() 메서드를 호출
                            #인자로 'list'를 입력하면 딕셔너리의 값이 리스트 형으로 변환된다.
print(dict3)
```

```python
# DataFrame에서 데이터의 삭제는 drop() 메서드를 이용한다.
# 기존 칼럼 값을 가공해 새로운 칼럼을 만들고 삭제하는 경우가 많다보니 주로 axis = 1 을 설정한다.
# axis = 0으로 설정하고 로우를 삭제하는 경우는 이상치를 삭제하는 경우이다.
# inplace = True로 설정하면 기존 DataFrame의 데이터를 삭제한다. 디폴트 값이 False
```

```python
titanic_reset_df = titanic_df.reset_index(inplace = False)
titanic_reset_df.head()
# 새로운 인덱스가 연속 숫자 형으로 할당되며 기존 인덱스는 ['index'] 칼럼으로
# reset_index()는 인덱스가 연속된 int 숫자형 데이터가 아닐 경우에 주로 사용
```

```python
value_counts = titanic_df['Pclass'].value_counts()
value_counts
```

```python
new_value_counts = value_counts.reset_index() # 시리즈에 reset_index()를 적용하면 DataFrame으로 변환
new_value_counts
```

```python
# reset_index()의 파라미터에 drop = True를 설정하면 기존 인덱스가 삭제된다.
new_value_counts = value_counts.reset_index(drop = True)
new_value_counts
```

```python
# DataFrame loc[] 연산자
# 유의할 점: loc[]에 슬라이싱 기호를 적용하면 종료값-1이 아니라 종료값 까지를 포함한다. -> 명칭 기반 인덱싱이기 때문
```

### Aggregation 함수, Groupby 적용

```python
# sort_values()는 SQL의 DRER BY와 유사하다.
titanic_sorted = titanic_df.sort_values(by = ['Name'])
titanic_sorted.head()
```

```python
titanic_sorted = titanic_df.sort_values(by=['Pclass','Name'], ascending = False)
titanic_sorted.head()
```

```python
titanic_df.count()
```

```python
titanic_df[['Age','Fare']].mean()
```

```python
titanic_groupby = titanic_df.groupby('Pclass').count() # groupby() 대상 칼럼을 제외한 몯느 칼럼에 해당 aggregation 함수 적용
titanic_groupby
```

```python
titanic_groupby = titanic_df.groupby('Pclass')[['PassengerId','Survived']].count()
titanic_groupby
```

```python
titanic_df.groupby('Pclass')['Age'].agg([max,min,len])
```

```python
#SQL : SELECT max(Age), sum(SibSp), avg(Fare) FROM titanic_table GROUP BY Pclass
#DataFrame groupby() :
agg_format = {'Age':'max','SibSp':'sum','Fare':'mean'}
titanic_df.groupby('Pclass').agg(agg_format)
```

### 결손 데이터(누락 데이터) 처리

```python
titanic_df.isna().sum() # sum() 호출시 내부적으로 True는 1로, False는 0으로 변환
```

```python
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000') # Cabin열의 NaN 값을 C000으로
                                                            # 실제 값을 변경하려면 반환값을 titanic_df['Cabin'] 이렇게 정해주거나
                                                            #역시 inplace = True로 설정
titanic_df.head(3)
```

```python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df
```

```python
titanic_df['Age'].isna().sum()
```

### apply lambda

```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name','Name_len']].head().sort_values(['Name_len'])
```

```python
# 나이가 15세 미만이라면 'Child', 그렇지 않으면 'Adult'로 구분하는 새로운 칼럼 'Child|Adult'
titanic_df['Child|Adult'] = titanic_df['Age'].apply(lambda x: 'Child' if x<=15 else 'Adult')
                                                    # lambda x : if x <=15 'Child', else 'Adult'가 아니다.
                                                    # lambda는 elif를 지원하지 않는다
titanic_df[['Age','Child|Adult']].head(10)
```

```python
# elif 적용 방법
titanic_df['Age_Cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ( 'Adult' if x<= 60 else 'Elderly' ))
titanic_df['Age_Cat'].value_counts()
```
